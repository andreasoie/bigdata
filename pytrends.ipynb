```python
def get_historical_dataframe(keyword, year_start = 2021, month_start=1, day_start=1, hour_start=1, year_end = 2021, month_end=12, day_end=1, hour_end=0, geo='', hourly=True):
    kw_list = [keyword]
    tf = str(year_start) + "-" + str(month_start) + "-" + str(day_start) + " " + str(year_end) + "-" + str(month_end) + "-" + str(day_end)
    pytrend = TrendReq(hl='en-US', tz=360, timeout=(10,25), retries=2, backoff_factor=0.1, requests_args={'verify':False})
    pytrend.build_payload(kw_list, timeframe=tf, geo=geo)
    # print(tf)
    
    # next we make the dataframe for this word over the period
    df = 1
    if hourly: # if we want data for every hour in that period
        df = pytrend.get_historical_interest(kw_list,year_start=year_start, month_start=month_start, day_start=day_start, hour_start=hour_start, year_end=year_end, month_end=month_end, day_end=day_end, hour_end=hour_end, cat=0, geo=geo, gprop='', sleep=0)
    else: # if we only want data per day
        df = pytrend.interest_over_time()
    
    return df.drop(columns=["isPartial"])

# Example
# kw = "pokemon"
# df_pokemon = get_historical_dataframe(kw,year_start=2020, month_start=10, day_start=1, hour_start=0, year_end=2021, month_end=1, day_end=1, hour_end=0, geo='', hourly=True)
# df_pokemon
```

```python
def get_trending_data(keywords,countries,year_start = 2021, month_start=1, day_start=1, hour_start=1, year_end = 2021, month_end=3, day_end=1, hour_end=0, hourly=True):
    search_data = []
    for c in countries:
        for kw in keywords:
            print("Searching for", kw, "in country:",c)
            df = get_historical_dataframe(kw, year_start=year_start, month_start=month_start, day_start=day_start, hour_start=hour_start, year_end=year_end, month_end=month_end, day_end=day_end, hour_end=hour_end, geo=c, hourly=hourly)
            df_list = df_to_2D_list(df)
            if c=='':
                search_data += [np.array(['GLOBAL',kw,df_list[0],df_list[1]])]
            else:
                search_data += [np.array([c,kw,df_list[0],df_list[1]])]
    return search_data


keywords = ['pokemon','minecraft']
countries = ['','NO']
data_test = get_trending_data(keywords,countries,hourly=False)
print(data_test)
```
